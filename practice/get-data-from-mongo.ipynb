```python
from pymongo import MongoClient

client = MongoClient() # Creates a client that uses the default port on localhost
mongodb = client.yelp_climbing_gyms # Connect to yelp_climbing_gyms database
```

# Get Data

```python
mongodb.list_collection_names()
```

```python
# Read review information into memory
review_fields = {'_id':0, 'review_id': 1, 'user_id':1, 'user_name':1, 'business_id':1,
                 'stars':1, 'useful':1, 'funny':1, 'cool':1, 'text':1}
reviews = mongodb.reviews.find({}, review_fields)
review_tuple = list(reviews)
review_tuple
```

```python
len(review_tuple)
```

```python
# Read business information into memory

business_fields = {'_id': 0, 'link_id': 1, 'name': 1,
                   'address': 1, 'city': 1, 'state': 1, 'postal_code': 1, 'telephone': 1,
                   'latitude': 1, 'longitude': 1,
                   'stars': 1, 'review_count': 1,
                   'is_open': 1, 'attributes': 1, 'categories': 1, 'hours':1}

business = mongodb.business.find({}, business_fields)
business_tuple = list(business)
business_tuple
```
