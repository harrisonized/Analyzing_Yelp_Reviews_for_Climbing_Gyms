```python
new_gyms_cbj_df = pd.read_csv('data/new-gyms-cbj.csv') # this dataset has labeled targets, an improvement on df above
```

```python
# Specify data
trace1 = go.Scatter(
    x=new_gyms_cbj_df['Year'],
    y=new_gyms_cbj_df["Total Gyms"],
    name='Total Gyms'
)

trace2 = go.Scatter(
    x=new_gyms_cbj_df['Year'],
    y=new_gyms_cbj_df["New Gyms"],
    name='New Gyms',
    yaxis='y2'
)

data = [trace1, trace2]



# Layout
layout = go.Layout(
    autosize = False,
    width = 900,
    height = 600,
    title = 'USA Commercial Climbing Gym Growth',
    titlefont = dict(
            size = 40
        ),
    xaxis = dict(
        title = 'Year',
        range = [2008.5, 2019.5],
        dtick = 1,
        titlefont = dict(
            size = 32,
        ),
        tickfont = dict(
            size = 16
        ),
    ),
    
    yaxis = dict(
        title = 'Total Gyms',
        range = [0, 590],
        dtick = 100,
        titlefont = dict(
            size = 32,
            color = '#1f77b4'
        ),
        tickfont = dict(
            size = 18,
            color = '#1f77b4'
        ),
    ),
    yaxis2 = dict(
        title = 'New Gyms',
        overlaying = 'y',
        range = [0, 70.8],
        dtick = 12,
        side = 'right',
        titlefont = dict(
            size = 32,
            color = '#ff7f0e'
        ),
        tickfont = dict(
            size = 18,
            color = '#ff7f0e'
        ),
    ),
    
    showlegend = False,
    legend = dict(
        xanchor = 'center',
        x = 0.75, 
        orientation = 'h',
        font = dict(
            size = 20
        )
    )
)



# Show figure
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='multiple-axes-double')
```
